```python
# Answer of 3
def largest_factor(num): 
   # returns the highest factor
    for i in range(num-1,0,-1): 
        if num % i == 0:       
            return i          
largest_factor(15)
```

```python
#Answer of 4
# here the function computes the factor of the argument passed
def print_factors(x):
   print("The factors of",x,"are:")
   for i in range(1, x + 1):
       if x % i == 0:
           print(i)

num = 7
print_factors(num)

```
